# 01. DATA COLLECTION

## 1. Lý do chọn đề tài

Cuối tháng 10, đầu tháng 11 năm 2025 — mùa mưa của Việt Nam. Thời điểm này vừa xảy ra trận lũ lịch sử tại Huế, gây thiệt hại nặng nề về người và tài sản Từ sự tò mò và mong muốn đóng góp vào công tác dự báo, phân tích thời tiết, nhóm quyết định thực hiện đề tài thu thập và phân tích dữ liệu thời tiết, trong đó hiện tượng mưa được chọn làm tâm điểm tìm hiểu.

## 2. Thu thập dữ liệu



### 2.1. Các vấn đề trong quá trình thu thập:


#### a. Phương pháp thu thập data:

Có ba hướng tiếp cận chính được xem xét:

- Gọi API (phương pháp được nhóm sử dụng):

    - Ưu điểm:

        + Dễ dàng triển khai, chỉ cần sử dụng các endpoint có sẵn.

        + Dữ liệu thu được thường có định dạng chuẩn (JSON, CSV, v.v.), thuận tiện cho xử lý sau này

    - Hạn chế:

        + Một số API yêu cầu trả phí hoặc giới hạn số lần truy cập trong ngày.

- Web scraping:

    - Ưu điểm:

        + Nguồn dữ liệu dồi dào từ các trang tin tức, tạp chí hoặc cổng thông tin thời tiết.

    - Hạn chế:

        + Cấu trúc HTML thường phức tạp, đòi hỏi quá trình làm sạch dữ liệu (data cleaning).

        + Tính tin cậy và độ nhất quán của dữ liệu phụ thuộc vào từng website.

- Tự sinh data:

    - Hạn chế: Không phù hợp với định hướng của nhóm là phân tích dựa trên dữ liệu thực tế, nên phương án này bị loại bỏ sớm.


#### b. Nguồn thu thập data:

- OpenWeatherMap: Cung cấp dữ liệu chất lượng cao, tuy nhiên gói miễn phí chỉ cho phép 1000 lượt gọi/ngày, yêu cầu đăng ký và khai báo thẻ tín dụng (Visa/MasterCard). Việc thu thập dữ liệu lớn (≥10.000 dòng) trở nên khó khả thi.

- WeatherAPI: Dễ đăng ký và hoàn toàn miễn phí. Tuy nhiên, API chỉ cung cấp dữ liệu dự báo (forecast), không có dữ liệu thời tiết thực tế (current), nên không phù hợp với mục tiêu của nhóm.

- NOAA (National Oceanic and Atmospheric Administration): Là nguồn dữ liệu thô (raw data) có độ chính xác cao, được nhiều nền tảng thời tiết sử dụng. Tuy nhiên, quy trình truy xuất phức tạp nên chỉ được xem là phương án dự phòng cuối cùng.

- Open-Meteo: Mặc dù là nguồn dữ liệu miễn phí, dễ truy xuất và có chất lượng rất cao, tuy nhiên phần lớn dữ liệu được xây dựng từ các mô hình tái phân tích (reanalysis) và mô hình dự báo. Do đó, nguồn này không đáp ứng yêu cầu của nhóm về việc chỉ sử dụng dữ liệu quan trắc thực tế, nên không được lựa chọn trong nghiên cứu này.

- Meteostat (nguồn dữ liệu nhóm sử dụng): Cung cấp dữ liệu thời tiết thực tế được quan trắc từ các trạm khí tượng trên toàn thế giới, cho phép kiểm soát và chọn lọc dữ liệu theo trạm, thời gian và loại chỉ số. Meteostat không phụ thuộc vào dữ liệu sinh ra từ các mô hình dự báo hay tái phân tích, nhờ đó phù hợp hơn với mục tiêu phân tích dữ liệu thực nghiệm và đảm bảo tính minh bạch của nguồn dữ liệu.


#### c. Loại dữ liệu:

Dữ liệu thời tiết thường được chia thành hai nhóm chính:

- History — phản ánh điều kiện thời tiết đã xảy ra.

- Forecast — trạng thái thời tiết dựa trên mô hình dự đoán.

Nhóm lựa chọn dữ liệu lịch sử thay vì dữ liệu dự báo, vì mục tiêu nghiên cứu tập trung vào việc phân tích và nhận diện hiện tượng thật đã xảy ra.


### 2.2. Nguồn dữ liệu: <u>[Meteostat](https://meteostat.net/en/)</u>

- Meteostat là nền tảng cung cấp dữ liệu thời tiết miễn phí, tập trung vào dữ liệu quan trắc thực tế được thu thập từ các trạm khí tượng trên toàn thế giới. Dữ liệu bao gồm các chỉ số thời tiết theo giờ hoặc theo ngày như nhiệt độ, lượng mưa, độ ẩm, tốc độ gió, áp suất và độ che phủ mây,... cho phép truy xuất dữ liệu lịch sử đã xảy ra trong quá khứ một cách trực tiếp và minh bạch.

- Không giống các nền tảng dựa trên dữ liệu tái phân tích hoặc mô hình dự báo, Meteostat không sử dụng dữ liệu sinh ra từ mô hình khí tượng mà tổng hợp từ nhiều nguồn trạm quan trắc uy tín (bao gồm NOAA, Environment Canada và các cơ quan khí tượng quốc gia). Nhờ đó, nguồn dữ liệu này đặc biệt phù hợp cho mục tiêu phân tích thời tiết thực tế tại Việt Nam cũng như đáp ứng mục tiêu đồ án môn học.

### 2.3. Quá trình thu thập:

a. Khảo sát và lựa chọn nguồn dữ liệu:

- Tiến hành tìm hiểu các nguồn dữ liệu thời tiết công khai, đánh giá tính sẵn có, độ tin cậy và định dạng dữ liệu trước khi lựa chọn Open-Meteo làm nguồn chính.

b. Thu thập dữ liệu qua API:

- Nhóm sử dụng Meteostat API để truy xuất dữ liệu thời tiết quan trắc theo giờ tại khu vực Thành phố Huế và các vùng lân cận, dựa trên hệ thống trạm khí tượng gần nhất.

- Các thông tin được thu thập bao gồm: thời gian quan trắc, nhiệt độ không khí, độ ẩm, lượng mưa, tốc độ gió, hướng gió, áp suất khí quyển và nhiệt độ điểm sương,... đáp ứng yêu cầu phân tích chi tiết theo thời gian.


c. Lưu trữ: 
- Dữ liệu sau khi thu thập được tổng hợp và chuyển đổi sang định dạng CSV, tạo thuận lợi cho quá trình tiền xử lý, trực quan hóa và phân tích thống kê trong các bước tiếp theo.

### 2.4. Các đặc trưng được chọn:

- latitude, longitude: cặp (vĩ độ, kinh độ)

- datetime: thời điểm dữ liệu được ghi lại theo giờ trong ngày.

- temperature_2m (độ C): nhiệt độ không khí được đo ở độ cao 2m so với mặt đất, trung bình theo giờ trong ngày.

- dewpoint_2m (độ C): nhiệt độ điểm sương, nhiệt độ hơi nước cần lạnh đi để đạt 100% độ ẩm, từ đó bắt đầu ngưng tụ.

- relativehumidity_2m (%): độ ẩm tương đối (lương hơi nước trong không khí / lượng tối đa không khí ở nhiệt độ đó có thể giữ).

- precipitation (mm): lượng mưa.

- windspeed_10m (km/h): tốc độ gió ở độ cao 10m so với mặt đất.

- winddirection_10m (độ): hướng gió.

- pressure_msl (hPa): áp suất tại mực nước biển trung bình, là chuẩn để phân biệt áp cao/thấp.

```python
from meteostat import Hourly, Stations
from datetime import datetime
import pandas as pd
import numpy as np
import time

# 7 tọa độ đại diện cho các khu vực trong TP Huế và vùng lân cận
coords = [
    (16.5800, 107.6100),  # Bắc trung tâm: Huyện Hương Trà
    (16.5200, 107.5900),  # Trung tâm TP Huế
    (16.4700, 107.6200),  # Nam trung tâm: Huyện Hương Thủy
    (16.4300, 107.5800),  # Tây trung tâm: Huyện Phong Điền
    (16.3800, 107.6200),  # Nam Tây: Huyện Quảng Điền
    (16.5300, 107.6700),  # Đông Nam: Huyện Phú Vang
    (16.4400, 107.6500)   # Nam: Huyện A Lưới
]


# Thời gian lấy dữ liệu
start = datetime(2021, 1, 1)
end   = datetime(2025, 11, 30)
output_file = "data/raw/weather_raw.csv"
all_data = []

for lat, lon in coords:
    print(f"Đang xử lý ở tọa độ ({lat}, {lon})")
    # Tìm trạm gần nhất
    stations = Stations().nearby(lat, lon)
    station = stations.fetch(1)

    if station.empty:
        print("Không tìm thấy trạm")
        continue

    station_id = station.index[0]
    print("Trạm:", station_id)

    # Lấy dữ liệu gốc, không lấy dữ liệu do model sinh ra
    data_real = Hourly(station_id, start, end, model=False)
    df_real = data_real.fetch()

    # Do cột prcp tất cả được sinh ra từ việc lấy tổng hợp các vùng lân cận và gom cụm thông qua model
    # Nên cột này thật chất vẫn là dữ liệu thật chứ không phải do mô hình dự đoán
    data_model = Hourly(station_id, start, end, model=True)
    df_model = data_model.fetch()[["prcp"]]

    if df_real.empty:
        print("Trạm có nhưng không có dữ liệu")
        continue

    # Ghi đè prcp vào
    df_real["prcp"] = df_model["prcp"]

    # Reset index
    df = df_real.reset_index()

    # Gắn tọa độ
    df["latitude"] = lat
    df["longitude"] = lon

    # Rename đúng cột 
    df = df.rename(columns={
        "time": "datetime",
        "temp": "temperature (C degree)",
        "rhum": "humidity (%)",
        "dwpt": "dewpoint (C degree)",
        "prcp": "precipitation (mm)",
        "wspd": "windspeed (km/h)",
        "wdir": "winddirection (degrees)",
        "pres": "pressure (hPa)"
    })

    # Chọn cột cần thiết
    df = df[
        ["latitude","longitude","datetime",
         "temperature (C degree)","humidity (%)","precipitation (mm)",
         "windspeed (km/h)","winddirection (degrees)",
         "pressure (hPa)","dewpoint (C degree)"]
    ]

    all_data.append(df)
    time.sleep(1)

# Gộp và ghi file 
if all_data:
    final_df = pd.concat(all_data, ignore_index=True)
    final_df.to_csv(output_file, index=False, encoding="utf-8")
    print(f"\nĐã lưu dữ liệu: {output_file}")
else:
    print("Không có dữ liệu")

final_df=pd.DataFrame(final_df)
print(final_df.shape)
```
